# Droping columns

```python
df.drop(['CCA3','Country/Territory','Capital','1990 Population','1980 Population', '1970 Population', 'Area (km²)', 'Density (per km²)',],axis=1,inplace=True)
```

```python
df
```

# Checking null value

```python
df.isna().sum()
```

# Checking unique Values

```python
df['Continent'].unique()
```

# Replacing value

```python
df.Continent=df.Continent.replace({'Asia':1,'Europe':2,'Africa':3,'Oceania':4,'North America':5,'South America':6})
```

```python
df
```

# Checking datatypes

```python
df.dtypes
```

# Converting datatypes to integer

```python
df['Growth Rate']=df['Growth Rate'].astype(int)
df['World Population Percentage']=df['World Population Percentage'].astype(int)
```

```python
df.dtypes
```

# Stastical information

```python
df.describe()
```

# Seprating Dependent and Indeoendent Variable

```python
x=df.drop(['Continent'],axis=1)
x
```

```python
y=df.Continent
y
```

# Importing train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier(n_estimators=100)
```

```python
model_rf.fit(x_train,y_train)
```

```python
model_rf.score(x_train,y_train)*100
```

```python
model_rf.score(x_test,y_test)*100
```

```python
y_predict=model_rf.predict(x_test)
```

```python
y_predict
```

```python
df=pd.DataFrame({'Actual':y_test,'Predicted':y_predict})
df
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_acc = accuracy_score(y_test,y_predict)*100
```

```python
test_acc
```

# Classification report

```python
from sklearn.metrics import classification_report
```

```python
performance=classification_report(y_test,y_predict)
```

```python
print(performance)
```
